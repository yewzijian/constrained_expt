```python
# ARC
success = [.8, .8, .6, .8, .8, .8]
algorithms = ('RRT', 'RRTConnect ', 'EST',  'RRT', 'RRTConnect', 'EST')

fig, ax = plt.subplots(figsize=(10,4))
h = ax.bar( [0.5,1,1.5,2.5,3,3.5], success, width=0.49, color=['b', 'b', 'b', 'g', 'g', 'g'])
ax.set_xticks( [0.5,1,1.5,2.5,3,3.5])
ax.set_xticklabels(algorithms)
ax.legend((h[0], h[3]), ('Joint space', 'Pose space'))
ax.set_ylabel('Succes rate')
ax.set_ylim([0,1])
ax.set_xlim([0,4])

fig.tight_layout()
plt.show()
```
